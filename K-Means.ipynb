```python
#He creado otra variable del dataset, para que no se mezcle con la primera parte de visualización de datos.
irisdf = iris
```

```python
#Eliminamos la columna "Species" del dataframe (el objetivo), y metemos el valor en la variable "y"
X = irisdf.drop("Species", axis=1)
y = irisdf["Species"]
```

```python
#dividimos en dos conjuntos; el de prueba y el de test (70%/30%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
#Preporcesamiento: escalado de datos
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
#Aplicamos K-means, con 3 clusters y el random state para que pueda volver a ser reproducido. Luego, realizamos el entrenamiento con los datos escalados.
kmeans = KMeans(n_clusters=3, random_state=1)
kmeans.fit(X_scaled)
```

```python
# Obtenemos las etiquetas de los clusters y las cordenadas de sus centroides.
labels = kmeans.labels_

print("Centroides de los clusters:")
print(kmeans.cluster_centers_)
```

```python
# Representamos gráficamente el cluster:
plt.scatter(X_scaled[:, 2], X_scaled[:, 3], c=labels, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 2], kmeans.cluster_centers_[:, 3], c='red', marker='x', s=200)  
plt.xlabel('Largo del pétalo (escalado)')
plt.ylabel('Ancho del pétalo (escalado)')
plt.title('K-Means Para el trabajo del dataset de Iris')
plt.colorbar(label='Cluster')
plt.show()
```

```python
inertia = kmeans.inertia_
silhouette = silhouette_score(X_scaled, kmeans.labels_)
```

```python
wandb.log({
    "inertia": inertia,
    "silhouette_score": silhouette,
    "num_clusters": 3
})
```
